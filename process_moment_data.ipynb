```python
import numpy as np
import matplotlib.pyplot as pp
import pandas as pd

def printGraph(df, df_bleed, startRow, startRow_bleed, color, color_bleed):
   val = 0. 
   #Average Data
   rowData = df.iloc[ startRow , 1:10]
   rowData_bleed = df_bleed.iloc[ startRow_bleed , 1:10]
   print('Average Graph')
   pp.plot(rowData, np.zeros_like(rowData) + val, 'x', color=color)
   pp.plot(rowData_bleed, np.zeros_like(rowData_bleed) + val, 'x', color=color_bleed)
   pp.show()
   
   #Stad Dev Data
   rowData = df.iloc[ startRow+1 , 1:10]
   rowData_bleed = df_bleed.iloc[ startRow_bleed+1 , 1:10]
   print('Stad Dev Graph')
   pp.plot(rowData, np.zeros_like(rowData) + val, 'x', color=color)
   pp.plot(rowData_bleed, np.zeros_like(rowData_bleed) + val, 'x', color=color_bleed)
   pp.show()
   
   #Skew Data
   rowData = df.iloc[ startRow+2 , 1:10]
   rowData_bleed = df_bleed.iloc[ startRow_bleed+2 , 1:10]
   print('Skew Graph')
   pp.plot(rowData, np.zeros_like(rowData) + val, 'x', color=color)
   pp.plot(rowData_bleed, np.zeros_like(rowData_bleed) + val, 'x', color=color_bleed)
   pp.show()
   
   #Kurtosis Data
   rowData = df.iloc[ startRow+3 , 1:10]
   rowData_bleed = df_bleed.iloc[ startRow_bleed+3 , 1:10]
   print('Kurtosis Graph')
   pp.plot(rowData, np.zeros_like(rowData) + val, 'x', color=color)
   pp.plot(rowData_bleed, np.zeros_like(rowData_bleed) + val, 'x', color=color_bleed)
   pp.show()
   
   #Max Data
   rowData = df.iloc[ startRow+4 , 1:10]
   rowData_bleed = df_bleed.iloc[ startRow_bleed+4 , 1:10]
   print('Max Graph')
   pp.plot(rowData, np.zeros_like(rowData) + val, 'x', color=color)
   pp.plot(rowData_bleed, np.zeros_like(rowData_bleed) + val, 'x', color=color_bleed)
   pp.show()
   
   #Min Data
   rowData = df.iloc[ startRow+5 , 1:10]
   rowData_bleed = df_bleed.iloc[ startRow_bleed+5 , 1:10]
   print('Min Graph')
   pp.plot(rowData, np.zeros_like(rowData) + val, 'x', color=color)
   pp.plot(rowData_bleed, np.zeros_like(rowData_bleed) + val, 'x', color=color_bleed)
   pp.show()
   return

data_file = r'C:\Users\Akash Verma\Documents\Dissertation\Research\Moments_data_normal.xlsx'
df = pd.read_excel(data_file)
data_file = r'C:\Users\Akash Verma\Documents\Dissertation\Research\Moments_data_bleeding.xlsx'
df_1 = pd.read_excel(data_file)
printGraph(df,df_1,101,121, 'blue','red')
```
